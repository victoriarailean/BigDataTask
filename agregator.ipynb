### Import dependencies

```python
import pandas as pd
```

### Parse csv files into dataframe

```python
facebook_ds = pd.read_csv(r"data/facebook_dataset.csv", on_bad_lines='skip')
google_ds = pd.read_csv(r"data/google_dataset.csv", on_bad_lines='skip')
website_ds = pd.read_csv(r"data/website_dataset.csv", on_bad_lines='skip', sep=";")
```

### Necessary fields

```python
common_columns = ["company_name", "categories", "country", "phone", "domain" , "city", "region_name"]

facebook_columns = common_columns + ["address"]
google_columns   = common_columns + ["address"]
website_columns  = common_columns
```

### Rename columns

```python
facebook_ds = facebook_ds.rename(columns={'name': 'company_name', 'country_name': 'country'})
facebook_ds
```

```python
google_ds = google_ds.rename(columns={'name': 'company_name', 'category': 'categories', 'country_name': 'country'})
google_ds
```

```python

website_ds = website_ds.rename(columns={'root_domain':'domain', 'main_country': 'country', 's_category': 'categories', 'legal_name':'company_name', 'main_city':'city', 'main_region':'region_name'})
website_ds
```

### Merge facebook data with google

```python
merged_df = pd.merge(facebook_ds[facebook_columns], google_ds[google_columns], on='domain', how='outer', suffixes=["_facebook", "_google" ])
merged_df

```

#### Assign values and remove temporary  columns

```python
merged_df['company_name'] = merged_df['company_name_facebook'].fillna(merged_df['company_name_google'])
merged_df['categories'] = merged_df['categories_facebook'].fillna(merged_df['categories_google'])
merged_df['country'] = merged_df['country_facebook'].fillna(merged_df['country_google'])
merged_df['phone'] = merged_df['phone_facebook'].fillna(merged_df['phone_google'])
merged_df['city'] = merged_df['city_facebook'].fillna(merged_df['city_google'])
merged_df['region_name'] = merged_df['region_name_facebook'].fillna(merged_df['region_name_google'])
merged_df['address'] = merged_df['address_facebook'].fillna(merged_df['address_google'])

dropped_columns = [
    'company_name_facebook', 
    'company_name_google', 
    'categories_facebook', 
    'categories_google', 
    'country_facebook', 
    'country_google', 
    'phone_facebook', 
    'phone_google', 
    'city_facebook', 
    'city_google',
    'region_name_facebook',
    'region_name_google',
    'address_facebook',
    'address_google'
]

merged_df.drop(columns=dropped_columns, inplace=True)
merged_df
```

### Merge the result with the web site date

```python

merged_web = pd.merge(merged_df[facebook_columns], website_ds[website_columns], on='domain', how='outer', suffixes=["_temp", "_website"])
merged_web
 
```

#### Assign values and remove temporary  columns

```python
merged_web['company_name'] = merged_web['company_name_website'].fillna(merged_web['company_name_temp'])
merged_web['categories'] = merged_web['categories_website'].fillna(merged_web['categories_temp'])
merged_web['country'] = merged_web['country_website'].fillna(merged_web['country_temp'])
merged_web['phone'] = merged_web['phone_website'].fillna(merged_web['phone_temp'])
merged_web['city'] = merged_web['city_website'].fillna(merged_web['city_temp'])
merged_web['region_name'] = merged_web['region_name_website'].fillna(merged_web['region_name_temp'])

dropped_columns = [
    'company_name_website', 
    'company_name_temp', 
    'categories_website', 
    'categories_temp', 
    'country_website', 
    'country_temp', 
    'phone_website', 
    'phone_temp',
    'city_website',
    'region_name_website',
    'region_name_temp',
    'city_temp'

    ]

merged_web.drop(columns=dropped_columns, inplace=True)

merged_web
```

### Remove duplicates 

```python
merged_web = merged_df.drop_duplicates().reset_index(drop=True)
```

### Export dataframe to csv

```python
merged_web.to_csv('data/merged_dataset.csv', index=False)
```
